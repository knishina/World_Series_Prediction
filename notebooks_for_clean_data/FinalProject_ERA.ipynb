```python
# Import dependencies.
import pandas as pd
from collections import Counter
```

# ERA

### Import 1905 and 1969 csvs and isolate the incidence of average ERA values for World Series winners and non-World Series Winners.

```python
# Open up the 1905.csv and inspect.
df2 = pd.read_csv("../clean_data/1905.csv")
df2 = df2.drop("Unnamed: 0", axis=1)
df2
```

```python
# Open up the 1969.csv and inspect.
df3 = pd.read_csv("../clean_data/1969.csv")
df3 = df3.drop("Unnamed: 0", axis=1)
df3
```

### Gather the winners of the World Series.
    # Collect data in list for ws_win ERAs.
    # Collect data in list for nonws_win ERAs.
    # Collect data for all ERAs.
    # Turn data into dataframes.

```python
# Make a list of World Series winners and non-World Series winners ERA for 1905 onwards.
ws_win_ERA_1905 = []
ws_nowin_ERA_1905 = []
for entry in range(len(df2)):
    if df2["WSWin"][entry] == "Y":
        ERA = str(round((df2["ERA"][entry]), 1))
        ws_win_ERA_1905.append(float(ERA))
    elif df2["WSWin"][entry] == "N":
        ERA = str(round((df2["ERA"][entry]), 1))
        ws_nowin_ERA_1905.append(float(ERA))
```

```python
# Make a list of World Series winners and non-World Series winners ERA for 1969 onwards.
ws_win_ERA_1969 = []
ws_nowin_ERA_1969 = []
for entry in range(len(df3)):
    if df3["WSWin"][entry] == "Y":
        ERA = str(round((df3["ERA"][entry]), 1))
        ws_win_ERA_1969.append(float(ERA))
    elif df3["WSWin"][entry] == "N":
        ERA = str(round((df3["ERA"][entry]), 1))
        ws_nowin_ERA_1969.append(float(ERA))
```

```python
# Make a list of all ERA for 1905 onwards.
ws_ERA_tot_1905 = []
for entry in range(len(df2)):
    ERA = str(round((df2["ERA"][entry]), 1))
    ws_ERA_tot_1905.append(float(ERA))
```

```python
# Make a list of all ERA for 1969 onwards.
ws_ERA_tot_1969 = []
for entry in range(len(df3)):
    ERA = str(round((df3["ERA"][entry]), 1))
    ws_ERA_tot_1969.append(float(ERA))
```

```python
# Find percent ERA frequency of World Series winners 1905 onwards.  Set up dataframe.
ws_win_counter_1905 = Counter(ws_win_ERA_1905)
wswin_1905 = pd.Series(ws_win_counter_1905, index = ws_win_counter_1905.keys())
wswin_1905 = wswin_1905.reset_index()
wswin_1905 = wswin_1905.rename(columns = {"index": "ERA", 0: "Count"})
```

```python
# Find percent ERA frequency of non-World Series winners 1905 onwards.  Set up dataframe.
ws_nowin_counter_1905 = Counter(ws_nowin_ERA_1905)
wsnowin_1905 = pd.Series(ws_nowin_counter_1905, index = ws_nowin_counter_1905.keys())
wsnowin_1905 = wsnowin_1905.reset_index()
wsnowin_1905 = wsnowin_1905.rename(columns = {"index": "ERA", 0: "Count"})
```

```python
# Find percent ERA frequency of World Series winners 1969 onwards.  Set up dataframe.
ws_win_counter_1969 = Counter(ws_win_ERA_1969)
wswin_1969 = pd.Series(ws_win_counter_1969, index = ws_win_counter_1969.keys())
wswin_1969 = wswin_1969.reset_index()
wswin_1969 = wswin_1969.rename(columns = {"index": "ERA", 0: "Count"})
```

```python
# Find percent ERA frequency of non-World Series winners 1969 onwards.  Set up dataframe.
ws_nowin_counter_1969 = Counter(ws_nowin_ERA_1969)
wsnowin_1969 = pd.Series(ws_nowin_counter_1969, index = ws_nowin_counter_1969.keys())
wsnowin_1969 = wsnowin_1969.reset_index()
wsnowin_1969 = wsnowin_1969.rename(columns = {"index": "ERA", 0: "Count"})
```

```python
# Find ERA frequency of all teams 1905 onwards.  Set up dataframe.
wins_ERA_counter_1905 = Counter(ws_ERA_tot_1905)
tot_ERA_1905 = pd.Series(wins_ERA_counter_1905, index = wins_ERA_counter_1905.keys())
tot_ERA_1905 = tot_ERA_1905.reset_index()
tot_ERA_1905 = tot_ERA_1905.rename(columns = {"index": "ERA", 0: "Count"})
```

```python
# Find ERA frequency of all teams 1969 onwards.  Set up dataframe.
wins_ERA_counter_1969 = Counter(ws_ERA_tot_1969)
tot_ERA_1969 = pd.Series(wins_ERA_counter_1969, index = wins_ERA_counter_1969.keys())
tot_ERA_1969 = tot_ERA_1969.reset_index()
tot_ERA_1969 = tot_ERA_1969.rename(columns = {"index": "ERA", 0: "Count"})
```

### Create another column for Frequency.
    # Turn the Count number into a frequency of the sample population.
    # Add the information to the dataframe.

```python
# Find frequency of World Series winners of 1905 onwards.
total1 = wswin_1905["Count"].sum()
frequency1 = [x/total1 for x in wswin_1905["Count"]]
wswin_1905["Frequency"] = frequency1
wswin_1905
```

```python
# Find frequency of non-World Series winners of 1905 onwards.
total2 = wsnowin_1905["Count"].sum()
frequency2 = [x/total2 for x in wsnowin_1905["Count"]]
wsnowin_1905["Frequency"] = frequency2
wsnowin_1905
```

```python
# Find frequency of World Series winners of 1969 onwards.
total1 = wswin_1969["Count"].sum()
frequency1 = [x/total1 for x in wswin_1969["Count"]]
wswin_1969["Frequency"] = frequency1
wswin_1969
```

```python
# Find frequency of non-World Series winners of 1969 onwards.
total2 = wsnowin_1969["Count"].sum()
frequency2 = [x/total2 for x in wsnowin_1969["Count"]]
wsnowin_1969["Frequency"] = frequency2
wsnowin_1969
```

```python
# Find frequency of all from 1905 onwards.
total5 = tot_ERA_1905["Count"].sum()
frequency5 = [x/total5 for x in tot_ERA_1905["Count"]]
tot_ERA_1905["Frequency"] = frequency5
tot_ERA_1905
```

```python
# Find frequency of all from 1969 onwards.
total5 = tot_ERA_1969["Count"].sum()
frequency5 = [x/total5 for x in tot_ERA_1969["Count"]]
tot_ERA_1969["Frequency"] = frequency5
tot_ERA_1969
```

### Export data for later analysis by bar chart.

```python
wswin_1905.to_csv("../clean_data/wswin_1905ERA.csv")
wsnowin_1905.to_csv("../clean_data/nowswin_1905ERA.csv")
wswin_1969.to_csv("../clean_data/wswin_1969ERA.csv")
wsnowin_1969.to_csv("../clean_data/nowswin_1969ERA.csv")
tot_ERA_1905.to_csv("../clean_data/totwins_1905ERA.csv")
tot_ERA_1969.to_csv("../clean_data/totwins_1969ERA.csv")
```
