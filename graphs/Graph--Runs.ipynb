## Perform basic stats analysis.
    # Calculate mean and standard deviation of Runs of the World Series winning population.
    # Calculate mean and standard deviation of Runs of the World Series non-winning population.
    # Calculate confidence interval.

### 1905.

```python
# Calculate mean of Runs of World Series winning teams for 1905 onwards.
wswin_mean1905 = round((np.mean(df2a["Runs"])), 2)
wswin_mean1905
```

```python
# Calculate standard deviation of World Series winning teams for 1905 onwards.
wswin_dev1905 = round((statistics.stdev(df2a["Runs"])), 2)
wswin_dev1905
```

```python
# Calculate number of entries in World Series winning teams for 1905 onwards.
wswin_len1905 = df2a["Count"].sum()
wswin_len1905
```

```python
# Calculate mean Runs fo non-World Series winning teams for 1905 onwards.
nowswin_mean1905 = round((np.mean(df2b["Runs"])), 2)
nowswin_mean1905
```

```python
# Calculate standard deviation of non-World Series winning teams for 1905 onwards.
nowswin_dev1905 = round((statistics.stdev(df2b["Runs"])), 2)
nowswin_dev1905
```

```python
# Calculate total number of non-World Series winning teams for 1905 onwards.
nowswin_len1905 = df2b["Count"].sum()
nowswin_len1905
```

### Determine significance of 1905 winners and non-winners.

```python
# Calculate variance for each sample group.
wswin_variance1905 = (wswin_dev1905)**2
nowswin_variance1905 = (nowswin_dev1905)**2
print (wswin_variance1905)
print (nowswin_variance1905)
```

    # The hypothesis I'm testing is that World Series winners tend to have higher Runs than non-World Series teams.
    # That means:
        # null hypothesis is:  u1 < u2....  or u1 - u2 <= 0
        # hypothesis to test is:  u1 - u2 > 0
    # This stands to be a right tail test.

```python
# Calculate Z.
num_1905 = wswin_mean1905 - nowswin_mean1905
dem_1905 = ((wswin_variance1905/wswin_len1905) + (nowswin_variance1905/nowswin_len1905)) ** 0.5
Z_1905 = num_1905/dem_1905
Z_1905
```

    # Because this is a right tail, find negative Z.  Then look up on table.

### Look up the Z value on table.  P-Value is < 0.00001
### This means that I can say that if the confidence level is 99.99% (alpha = 0.0001) that World Series teams definitely has a lower Runs than all other teams.... because p<= alpha (reject null hypothesis; accept test hypothesis).
